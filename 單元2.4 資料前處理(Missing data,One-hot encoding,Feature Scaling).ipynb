```python
import numpy as np
import pandas as pd
from sklearn import datasets
from io import StringIO
```

```python
csv_data = '''A,B,C,D,E
              5.5,2.0,3.0,,6
              1.0,6.0,,8.0,5
              0.0,11.0,12.0,4.0,5
              3.0,,3.0,5.0,
              5.0,1.0,4.0,2.0,4
           '''
```

```python
df = pd.read_csv(StringIO(csv_data))
```

```python
df
```

Missing Data(空值資料處理)

```python
#只要有空值整行皆去除
df.dropna()
```

```python
#全部都有空值才刪除
df.dropna(how='all')
```

```python
#指定某一欄位值為空的，就丟掉
df.dropna(subset=['C'])
```

```python
#填補缺失值，將缺失值皆填寫為0
df.fillna(0)
```

```python
#在B欄位中有缺失值，使用平均值做填補
df['B'] = df['B'].fillna(df['B'].mean())
df
```

```python
#在C欄位中有缺失值，使用眾數做填補
df['C'] = df['C'].fillna(df['C'].mode())
df
```

```python
#在D欄位中有缺失值，使用中位數做填補
df['D'] = df['D'].fillna(df['D'].median())
df
```

```python
#在E欄位中有缺失值，使用最小值做填補
df['E'] = df['E'].fillna(df['E'].min())
df
```

Categorical Data 類別資料的處理(有序、無序)

```python
df2 = pd.DataFrame([['green','M',10.1,1],
                    ['red','L',13.5,2],
                    ['blue','XL',15.3,1]])
df2.columns = ['color','size','price','classlabel']
df2
```

```python
#將size轉成數值
size_mapping = {'XL':3,
                'L':2,
                'M':1}
df2['size'] = df2['size'].map(size_mapping)
df2
```

```python
#One-hot encoding
pd.get_dummies(df2['color'])
```

```python
#prefix欄位名稱
onehot_encoding = pd.get_dummies(df2['color'],prefix='color')
```

```python
df2 = df2.drop('color',1)
df2
```

```python
pd.concat([onehot_encoding, df2],axis=1)
```

資料正規化、區間縮放(normallzation)

0-1區間縮放

```python
from IPython.display import Math
```

```python
Math(r'x^{(i)}_{norm}=\frac{x^{(i)}-x_{min}}{x_{max}-x_{min}}')
```

```python
iris = datasets.load_iris()
x = pd.DataFrame(iris['data'],columns = iris['feature_names'])
print("target_names: "+str(iris['target_names']))
y = pd.DataFrame(iris['target'],columns=['target'])
data = pd.concat([x,y],axis=1)
data.head(3)
```

```python
#對花萼長度做正規化
#\為換行
data['sepal length (cm)'] = (data['sepal length (cm)'] -data['sepal length (cm)'].min())/\
                            (data['sepal length (cm)'].max() - data['sepal length (cm)'].min())
data
```

資料標準化(Standardization)

資料減去平均值除標準差，平均變為0，標準差變為1，可以符合常態分佈

```python
Math(r'x^{(i)}_{std}=\frac{x^{(i)}-\mu_{x}}{\sigma_{x}}')
```

```python
#花萼寬度做正規化
data['sepal width (cm)'] = (data['sepal width (cm)'] - data['sepal width (cm)'].mean())/\
                           (data['sepal width (cm)'].std())
data.head()
```
