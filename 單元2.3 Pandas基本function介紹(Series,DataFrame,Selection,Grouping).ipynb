```python
from sklearn import datasets
import pandas as pd
import numpy as np
```

```python
iris = datasets.load_iris()
iris
```

```python
"target_names: "+str(iris['target_names'])
```

```python
x = pd.DataFrame(iris['data'],columns=iris['feature_names'])
x
```

```python
y = pd.DataFrame(iris['target'],columns=['target'])
y
```

```python
#合併欄位，水平合併
data = pd.concat([x,y],axis=1)
data
```

Data Calculation(資料運算)

```python
#pandas資料運算常用函式
data.apply(lambda x: x+1).head(5)
```

```python
#對整個表格做運算
data['sepal length (cm)'].apply(lambda x: x+2).head(5)
```

```python
#只能對指定column做運算
data['sepal length (cm)'].map(lambda x: x+2).head(5)
```

Data Grouping(資料分群)

```python
data.head()
```

```python
#每一個群體內的加總
data.groupby(by='target').sum()
```

```python
data.groupby(by='target').mean()
```

```python
for i in data.groupby(by='target'):
    print(i)
```

```python
#根據花萼長度做排序，取出第二大的資料
data.groupby(by='target').apply(lambda x:x.sort_values(by=['sepal length (cm)'], ascending=False)[1:2])
```

```python
#驗證上面的涵式是否正確
data[data['target']==0]['sepal length (cm)'].sort_values()
```
